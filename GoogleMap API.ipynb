```python
# Enter your Google Maps API key here
API_KEY = 'AIzaSyAilKSgEZ8oN6LhN-GlMrdtFqr2MjwDXjs'

# Define the search query and location
query = 'jewellery stores in birmingham'
location = 'Birmingham, UK'

# Set up the API request URL
url = f'https://maps.googleapis.com/maps/api/place/textsearch/json?query={query}&location={location}&radius=5000&key={API_KEY}'

# Send the API request and get the JSON response
response = requests.get(url)
data = response.json()

# Initialize an empty list to store the results
results = []

# Loop through the results and extract the location information
for result in data['results']:
    name = result['name']
    address = result['formatted_address']
    lat = result['geometry']['location']['lat']
    lng = result['geometry']['location']['lng']
    
    # Append the results to the list
    results.append({'Name': name, 'Address': address, 'Latitude': lat, 'Longitude': lng})

# Create a Pandas DataFrame from the list of results
df = pd.DataFrame(results, columns=["Name", "Address", "Latitude", "Longitude"])
df
```

```python
# Create a gmaps figure
gmaps.configure(api_key=API_KEY)
fig = gmaps.figure(center=(52.489471, -1.898575), zoom_level=12)

# Create a list to store marker objects and location coordinates
marker_list = []
location_list = []

# Create a layer for the jewelry stores
for index, row in df.iterrows():
    name = row['Name']
    address = row['Address']
    location = (row['Latitude'], row['Longitude'])
    
    # Create a marker for the store
    info_box_content = f"<p><b>{name}</b></p><p>{address}</p>"
    marker = gmaps.Marker(location, info_box_content=info_box_content)
    
    # Add the marker and location to the respective lists
    marker_list.append(marker)
    location_list.append(location)

# Create a marker layer for the store markers(nodes)
marker_layer = gmaps.Markers(markers=marker_list)
fig.add_layer(marker_layer)

# Create a layer for the links connecting the locations
line_list = []
for i in range(len(location_list)):
    for j in range(i+1, len(location_list)):
        start_location = location_list[i]
        end_location = location_list[j]
        line = gmaps.Line(start=start_location, end=end_location)
        line_list.append(line)

# Create a line layer for the connecting lines
line_layer = gmaps.drawing_layer(features=line_list)
fig.add_layer(line_layer)

# Display the figure
fig
```
